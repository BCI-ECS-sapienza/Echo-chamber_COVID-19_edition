```
import csv
lis = []
with open('/content/drive/My Drive/covid_project/graph_files/tweets_rows_deleted.csv', 'r', encoding='utf-8') as inp:
  for row in csv.reader(inp):
    lis.append(row[1])
```

```
#creating dictionary
from nltk import FreqDist
word_dist = FreqDist()
for s in lis:
    word_dist.update(s.split())
dic = dict(word_dist)
print(dic)
```

```
#ordering dictionary
dic_items = dic.items()
sorted_dic = sorted(dic_items)
print(sorted_dic)
```

```
with open('/content/drive/My Drive/covid_project/graph_files/dictionary.csv', 'w') as csv_file:  
    writer = csv.writer(csv_file)
    for elem in sorted_dic:
        writer.writerow([elem[0],elem[1]])
csv_file.close()
```

```
bst = []

values = sorted(dic.values())

for elem in range(1,20):
  i = - elem
  bst.append(values[i])
print(bst)
```

```
def getKeysByValue(dictOfElements, valueToFind):
    listOfKeys = list()
    listOfItems = dictOfElements.items()
    for item  in listOfItems:
        if item[1] == valueToFind:
            listOfKeys.append(item[0])
    return  listOfKeys
```

```
words = []
for elem in bst:
  words.append(getKeysByValue(dic,elem))
```
